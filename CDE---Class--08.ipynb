```python
url='https://m.timesjobs.com/mobile/jobs-search-result.html?txtKeywords=python&cboWorkExp1=-1&txtLocation='
web = requests.get(url).content
soup = BeautifulSoup(web,'html.parser')
titles=soup.find_all('div',class_='srp-job-heading')
```

```python
for title in titles:
    print(title.text)
```

```python
for title in titles:
    title=title.find('h3')

    print(title.text)
```

```python
jobs=soup.find_all('div',class_='srp-job-heading')
```

```python
job_json=[]
for job in jobs:
    title=job.find('h3').text
    company=job.find('span',class_='srp-comp-name').text
    posting_time=job.find('span',class_='posting-time').text
    lst = {title,company,posting_time}
    job_json.append(lst)
```
