```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
import time
import csv

# --- Set up Chrome driver with Service (headless optional) ---
service = Service("C:\\chromedriver-win32\\chromedriver.exe")  # Update the path to your chromedriver
options = Options()
options.add_argument("--start-maximized")
# options.add_argument("--headless")  # Uncomment for headless mode
driver = webdriver.Chrome(service=service, options=options)

# --- Go to CoinMarketCap ---
driver.get("https://coinmarketcap.com/")

# --- Scroll until 100 rows are loaded ---
max_scrolls = 30
scroll_pause = 2
last_count = 0

for i in range(max_scrolls):
    rows = driver.find_elements(By.XPATH, '//tbody/tr')
    print(f"Scroll {i+1}: {len(rows)} rows loaded")

    if len(rows) >= 100:
        break

    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(scroll_pause)

    if len(rows) == last_count:
        break
    last_count = len(rows)

rows = driver.find_elements(By.XPATH, '//tbody/tr')
print(f"\n✅ Final row count: {len(rows)}")

# --- Prepare to write CSV ---
with open('coinmarketcap_top_100.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow([
        "Rank", "Name", "Price", "1h Change", "24h Change", "7d Change",
        "Market Cap", "Volume"
    ])

    # --- Extract Data ---
    for i, row in enumerate(rows):
        try:
            cols = row.find_elements(By.TAG_NAME, 'td')
            
            # Ensure there are enough columns
            if len(cols) < 8:
                print(f"❌ Row {i+1} skipped due to missing columns.")
                continue

            rank = cols[0].text.strip()
            name = cols[1].text.strip().replace("\n", " ")
            price = cols[2].text.strip()
            change_1h = cols[3].text.strip()
            change_24h = cols[4].text.strip()
            change_7d = cols[5].text.strip()
            market_cap = cols[6].text.strip()
            volume = cols[7].text.strip()

            writer.writerow([
                rank, name, price, change_1h, change_24h, change_7d,
                market_cap, volume
            ])
        except Exception as e:
            print(f"❌ Error processing row {i+1}: {e}")
            continue

# --- Done ---
driver.quit()
print("\n✅ Data saved to 'coinmarketcap_top_100.csv'")
```
